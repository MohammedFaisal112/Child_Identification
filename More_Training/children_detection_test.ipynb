```python
import os
```

```python
training_fldr = 'train/adults'
training_fldr_2 = 'train/children'
```

```python
file_1 = os.listdir(training_fldr)
file_2 = os.listdir(training_fldr_2)
```

```python
import cv2 as cv2
```

```python
images = []
for i in file_1:
    total = training_fldr+'/'+i
    print(total)
    temp = cv2.imread(total)
    temp = cv2.cvtColor(temp , cv2.COLOR_BGR2RGB)
    temp = cv2.resize(temp,(48,48))
    images.append(temp)

```

```python
for i in file_2:
    total = training_fldr_2+'/'+i
    print(total)
    temp = cv2.imread(total)
    temp = cv2.cvtColor(temp , cv2.COLOR_BGR2RGB)
    temp = cv2.resize(temp,(48,48))
    images.append(temp)
```

```python
# Now Dependent Parameter extraction.
classification = []
for file in file_1:
    temp = file.split('_')[0]
    classification.append(temp)
```

```python
for file in file_2:
    temp = file.split('_')[0]
    classification.append(temp)
```

```python
converted_label = [0 if label == 'adult' else 1 for label in classification]
```

```python
import numpy as np
image_f = np.array(images)
label_f = np.array(converted_label)
```

```python
image_n = image_f/255.0
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test ,Y_train, Y_test = train_test_split(image_n,label_f,test_size=0.25)
```

```python
import tensorflow as tf
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
# from tensorflow.keras 
# import keras.api._v2.keras as keras
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dense , MaxPooling2D , Conv2D
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input , Activation , Add
from tensorflow.keras.optimizers import Adam , Adadelta , Adamax,Adagrad  , RMSprop
from tensorflow.keras.regularizers import l2 
```

```python
def Convolution(input_tesnor , filters):
    # Given filteration,kernal_size,padding,strides,kernel_regularizer=learning_rate=0.001
    x = Conv2D(filters=filters , kernel_size=(3,3) , padding='same' , strides=(1,1),kernel_regularizer=l2(0.001))(input_tesnor)
    # To reslove the overfitting problem use Dropout.
    x = Dropout(0.1)(x)
    # Activate the x
    x = Activation('relu')(x)
    return x
```

```python
# Defining Convolution Layers.
def model(input_shape):
    inputs = Input((input_shape))
    # Convolution layers:If more layers it'll create a biasness.If less layer it'll gives less accuracy.
    # So here taken 4 convolution layer.In testing whatever needed we will fixed with that.
    # 32 indicates a no.of filters.There's no such a fixed no.of filters.In image classification and object detection we will use state of art algorithm(currently best filters).
    conv1 = Convolution(inputs , 32)
    maxpool_1 = MaxPooling2D(pool_size=(2,2))(conv1)
    conv2 = Convolution(maxpool_1,64)
    maxpool_2 = MaxPooling2D(pool_size=(2,2))(conv2)
    conv3 = Convolution(maxpool_2 , 128)
    maxpool_3 = MaxPooling2D(pool_size=(2,2))(conv3)
    conv4 = Convolution(maxpool_3 , 256)
    maxpool_4 = MaxPooling2D(pool_size=(2,2))(conv4)
    flatten = Flatten()(maxpool_4)

    dense_1 = Dense(64,activation='relu')(flatten)
    drop_1 = Dropout(0.2)(dense_1)

    output_1 = Dense(1,activation='sigmoid',name = 'child_out')(drop_1)
    model = Model(inputs=[inputs], outputs = [output_1])
    model.compile(loss=['binary_crossentropy','mse'],optimizer='Adam',metrics=['accuracy'])
    return model
```

```python
# 48x48: Image width and height.
# 3: Number of color channels (RGB) for the colored image.
Model = model((48,48,3))
```

```python
Model.summary()
```

```python
# Importing Model checkpoints:
from keras.callbacks import ModelCheckpoint
```

```python
file_s = 'Child_Detection_test.h5'
# Monitoring=validation_loss
checkpoint = ModelCheckpoint(file_s,monitor='val_loss',verbose=1,save_best_only=True,save_weights_only=False,mode='auto',save_freq='epoch')
Early_stop = tf.keras.callbacks.EarlyStopping(patience=15,monitor='val_loss',restore_best_weights='True')
callback_list = [checkpoint,Early_stop]
```

```python
# Train the model.
# Model Started Learning out here.
# batch_size=64 is a one of the hyperparameter that can be tunned
History  = Model.fit(X_train,Y_train,batch_size=64,validation_data=(X_test,Y_test),epochs=100,callbacks=callback_list)
```
